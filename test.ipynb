```python
before = pd.read_csv("Before_Covid.csv")
count_u = 0
count_l = 0
i = 0
while i < len(before):
    if (before.loc[i, 'open'] - before.loc[i, 'close']) < 0:
        count_l +=1
    else: count_u += 1
    i += 1

vals = [count_u, count_l]
labels = ["Up " , "Down "]
fig, ax = plt.subplots()
ax.pie(vals, labels=labels)
ax.axis("equal")
```

```python
during = pd.read_csv("During_Covid.csv")
count_u = 0
count_l = 0
count_i = 0
i = 0
while i < len(during):
    if during.loc[i, 'open'] > max(during.loc[len(during)-1, 'open'], during.loc[0, 'open']):
        count_u += 1
    elif during.loc[i, 'open'] > min(during.loc[len(during)-1, 'open'], during.loc[0, 'open']):
        count_i += 1
    else: 
        count_l += 1
    i += 1

vals = [count_u, count_i, count_l]
labels = ["High outliers ", "Average Values ", "Low outliers "]
fig, ax = plt.subplots()
ax.pie(vals, labels=labels)
ax.axis("equal")
```
