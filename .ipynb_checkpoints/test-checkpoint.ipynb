```python
before = pd.read_csv("Before_Covid(2).csv")
during = pd.read_csv("During_Covid(4).csv")
after = pd.read_csv("After_Covid(2).csv")
for i in range(len(after)):
    after.loc[i, 'Source'] = after.loc[i, 'Source'][11:18]
```

```python
prev = after.loc[0, 'Source']
k = 1
i = 0
while (i < len(after)):
    while i < len(after) and prev == after.loc[i, 'Source']:
        if k < 10:
            after.loc[i, 'Source'] += '-0' + str(k)
        else:
            after.loc[i, 'Source'] += '-' + str(k)
        k += 1
        i += 1
    if i != len(after):
        prev = after.loc[i, 'Source']
        print(prev)
        if len(after.loc[i, 'Source']) == 7:
            after.loc[i, 'Source'] += '-01'
    i+=1
    k = 2
after[['Source', 'open', 'high', 'low', 'close', 'volume']].to_csv("After_Covid.csv", index = False)
```
