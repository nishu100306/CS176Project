```python
# Before covid outliers



# You might have to install these imports locally first!
import pandas as pd

# Load the CSV into a pandas DataFrame
df = pd.read_csv('Before_Covid.csv')

# Function to remove outliers using IQR method
def remove_outliers_iqr(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# List of numeric columns to check for outliers
numeric_columns = ['open', 'high', 'low', 'close', 'volume']

# Apply the IQR method to filter out outliers
df_filtered = remove_outliers_iqr(df, numeric_columns)

# Inspect the filtered DataFrame
print(df_filtered)

df_filtered.to_csv('Before_Covid_Outliers.csv', index=False)


```

```python
# Before covid boxplot




# You might have to install these imports locally first!
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the original and filtered CSV files
df_original = pd.read_csv('Before_Covid.csv')
df_outliers = pd.read_csv('Before_Covid_Outliers.csv')

def count_outliers_iqr(df, columns):
    outliers_count = {col: 0 for col in columns}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Count outliers
        outliers_count[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    return outliers_count

# List of numeric columns to check for outliers
numeric_columns = ['open', 'high', 'low', 'close', 'volume']

# Count outliers in the original and filtered datasets
original_outliers = count_outliers_iqr(df_original, numeric_columns)
filtered_outliers = count_outliers_iqr(df_outliers, numeric_columns)

# Print outliers counts
print("Original Outliers:", original_outliers)
print("Filtered Outliers:", filtered_outliers)

# Plotting boxplots for both original and filtered data

# Create a combined DataFrame for plotting
df_original_melted = pd.melt(df_original[['open', 'high', 'low', 'close', 'volume']], var_name='Variable', value_name='Value')
df_filtered_melted = pd.melt(df_filtered[['open', 'high', 'low', 'close', 'volume']], var_name='Variable', value_name='Value')

# Add a column to distinguish between original and filtered
df_original_melted['Dataset'] = 'Original'
df_filtered_melted['Dataset'] = 'Filtered'

# Combine both datasets for easy comparison
df_combined = pd.concat([df_original_melted, df_filtered_melted])

# Create a boxplot for the combined data
plt.figure(figsize=(12, 8))
sns.boxplot(x='Variable', y='Value', hue='Dataset', data=df_combined)

# Add title and labels
plt.title('Boxplot Comparison of Original vs Filtered Data (Before Covid)', fontsize=16)
plt.ylabel('Price/Volume', fontsize=12)
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
